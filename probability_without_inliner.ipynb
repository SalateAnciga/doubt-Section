```python
def get_pdf_prob(dataset,startrange,endrange):
    #Pyplot functionsused for creating a figure, plotting lines, and decorating the plot with labels.
    from matplotlib import pyplot
    #SciPy provides functions for working with the normal distribution,computing probability density functions (PDF), cumulative distribution functions (CDF)
    from scipy.stats import norm
    #Seaborn is a Python data visualization library built on top of Matplotlib. 
    import seaborn as sns
    #The distplot function  is used for univariate distribution of observations.kde means kernel distibution providing a smooth curve that represents the distribution of the data.
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    #Draw verticalline based on startrange,endrange with red color
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    #generate sample
    sample=dataset
    #calculator parameter
    #Calculate mean and standard Divation
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean-%.3f,Standard Deviation-%.3f'%(sample_mean,sample_std))
    #define distribution
    #norm() functionis used to make the  distribution
    dist=norm(sample_mean,sample_std)
    #sample probabilites for arange of outcomkes
    #create a list usind inliner for loop 
    values=[]
    for i in range(60,70):
        values.append(i)
    #find probability distribution
    probabilites=[dist.pdf(i) for i in values]
    #find whole probabilities using sum()
    prob=sum(probabilites)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilites)))
    return prob
```

```python
get_pdf_prob(dataset["ssc_p"],60,70)
```
